```python
from sklearn.datasets import load_iris
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
data = load_iris()
```

```python
df = pd.DataFrame(data.data, columns=data.feature_names)
```

```python
data.target
```

```python
df["label"] = data.target
df
```

```python
df.info()
```

```python
df.describe()
```

```python
#EDA

sns.pairplot(data=df, kind="scatter", hue="label", palette="Set2")
```

```python
df.isna().any()
```

```python
#ML MODELLING 
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(random_state=42, max_depth=5,min_samples_split=2,min_samples_leaf=2)
```

```python
X = df.drop("label",axis=1)
y=df.label
```

```python
#X_y splitting
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,stratify=y,random_state=42)
```

```python
rfc.fit(X_train,y_train)
```

```python
y_pred = rfc.predict(X_test)
```

```python
dir(rfc)
```

```python
from matplotlib import pyplot as plt

sns.barplot( x=rfc.feature_names_in_, y=rfc.feature_importances_)
plt.title("Feature_Importances of the data to the model!")
plt.show()
```

```python
y_pred 
```

```python
#Scikit metrics
from sklearn.metrics import classification_report, precision_recall_curve,confusion_matrix
```

```python
print(classification_report(y_pred,y_test))
```

```python
sns.heatmap(confusion_matrix(y_pred,y_test),annot=True,cmap="plasma")
```
